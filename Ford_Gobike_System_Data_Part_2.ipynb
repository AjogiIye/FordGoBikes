## Data Assessing and Cleaning

#### 1 .Drop null values in start_station_id,start_station_name,end_station_id,end_station_name,member_birth_year,member_gender

#### 2 .Convert Datatype of member_birth_year from float to int

#### 3 .Change start_time and end_time from object to datetime

#### 4 .Change to categorical variable user_type,member_gender,bike_share_share_for_all_trip

#### 5 .Create new column for day of the week and time of the day

#### 6 .Convert Datatype of duration_sec from integer to minutes

#### 7 .Convert member_birth_year to user year

```python
df_fordgobike = fordgo_bike.copy()
```

```python
# drop null values in 1 above
df_fordgobike.dropna(inplace=True)
```

```python
# test for null
df_fordgobike.isnull().sum()
```

```python
# generate function to test Dtype
def test_type(x):
    for var in x:
        print(f'{var} ----> {df_fordgobike[var].dtype}')
```

```python
# convert usertype,member_gender,bike_share_for_all_trip to categorical type
cass =['user_type', 'member_gender', 'bike_share_for_all_trip']
for var in cass:
    df_fordgobike[var]=df_fordgobike[var].astype('category')
```

```python
# conversion test
test_type(cass)
```

```python
# convert to datetime format start_time and end_time
cass =['start_time', 'end_time']
for var in cass:
    df_fordgobike[var]=pd.to_datetime(df_fordgobike[var])
    
```

```python
# test convertion
test_type(cass)
```

```python
# change birth year to integer 
df_fordgobike.member_birth_year=df_fordgobike.member_birth_year.astype('int')
```

```python
# remove and save in new column am and pm
df_fordgobike['time_of_day']=pd.to_datetime(df_fordgobike.start_time).dt.strftime('%p')
```

```python
# test column
df_fordgobike.time_of_day
```

```python
# remove day of the week to 'day_of_week' column
df_fordgobike['day_of_week']=pd.to_datetime(df_fordgobike.start_time).dt.strftime('%A')
```

```python
# test column
df_fordgobike.day_of_week
```

```python
# rename duration_sec to duration_min
def renameColumn(old_name,new_name):
    df_fordgobike.rename(columns={old_name:new_name},inplace=True)
# convert column name
renameColumn('duration_sec','duration_min')
df_fordgobike.duration_min=df_fordgobike.duration_min/60
```

```python
# test convertion
df_fordgobike.duration_min
```

```python
# retitle member_birth_year to user_age
renameColumn('member_birth_year','user_age')
# change birth_year to age
df_fordgobike.user_age=df_fordgobike.start_time.dt.year - df_fordgobike.user_age
```

## Duration minutes Spreading

1min to around 1000mins is the range for duration minutes and it peaked around 8mins.There emerged a unimodal shape of spreading when plotted on logarithic scale.

```python
# log scal plotting
def plot_hist(data,size,var,log_plot=False):
    if log_plot:
        plt.figure(figsize=[9,6]);
        bins=10**np.arange(np.log10(data['duration_min'].min()),np.log10(data['duration_min'].max())+size,size)
        plt.hist(data=data,x=var,bins=bins);
        plt.xscale('log');
        
    else:
        plt.figure(figsize=[9,6]);
        bins=np.arange(data[var].min(),data[var].max()+size,size)
        plt.hist(data=data,x=var,bins=bins);
        
plot_hist(df_fordgobike,0.06,'duration_min',log_plot=True)
plt.xlabel('minute consumed');
plt.ylabel('frequency');
plt.title('Duration Minutes Spreading');
xloc=[1,2,5,10,20,50,100,200,500,1000]
xlabel=['{}'.format(v) for v in xloc]
plt.xticks(xloc,xlabel);
```

## Categorical Variables Total Count Spreading

Count spreading presents the categories that are greater in number

```python
# generate age_category column for explicit spreading
df_fordgobike['age_category'] = pd.cut(x=df_fordgobike['user_age'], bins=[1, 20, 40, 60, 80, 100])
yloc=[20000,40000,60000,80000,100000,120000,140000,160000]
ylabel=['20k','40k','60k','80k','100k','120k','140k','160k']
# graph plotting
plot_title='Count Spreading for Member_Gender,Age_Category, and User_Type'

def plot_count(data,variables,title):
    fig, ax = plt.subplots(ncols=2, nrows=1, figsize=[20,8])
    fig.suptitle(title);
    color=sb.color_palette()[0]
    if len(variables) < 3:
        order={variables[0]:data[variables[0]].value_counts().index,
               variables[1]:data[variables[1]].value_counts().index,}
    else:
        order={variables[0]:data[variables[0]].value_counts().index,
               variables[1]:data[variables[1]].value_counts().index,
               variables[2]:data[variables[2]].value_counts().index,}
    count=1
    for var in variables:
        plt.subplot(1,3,count)
        sb.countplot(data=data,x=var,color=color,order=order[var]);
        plt.xticks(rotation=15)
        plt.ylabel('');
        plt.yticks(yloc,ylabel);
        count+=1
        
        
variables=['member_gender','age_category','user_type']

plot_count(df_fordgobike,variables,plot_title)
```

## Plot of Cycling Duration Verses User_Age

The plot below depicts a great concentration of cycling activity carried out by age range of 20years to around 60years. This implies that this age range is the most active of them all.

```python
plt.figure(figsize=[9,7])
sb.regplot(data=df_fordgobike,x='user_age',y='duration_min',fit_reg=False,scatter_kws={'alpha':1/3});
plt.yscale('log')
y_loc=[1,2,5,10,20,50,100,200,500,1000]
y_labels=['{}'.format(v) for v in y_loc]
plt.yticks(y_loc,y_labels);
plt.ylabel('Cycling Time');
plt.title('Plot of Cycling Duration Vs User Age');
plt.show();
```

## User_Type,Age_Category,Gender and Time_of_Day Total and Mean Cycling Time

Below are two dissimilar set of plots depicts that the group of age range (20 to 40)years old subscriber Males are greater in number but on average cycling , they spend smaller time.

### Spreading by count

```python
variables = ['age_category','member_gender','time_of_day']
count=1
fig, ax = plt.subplots(ncols=len(variables), figsize=[22,10])
for var in variables:
    plt.subplot(1,3,count)
    sb.countplot(data=df_fordgobike, x=var, hue='user_type');
    plt.xticks(rotation=90)
    plt.ylabel('')
    plt.legend(framealpha=0.5,title='User Type');
    yloc=[20000,40000,60000,80000,100000,120000,140000,160000]
    y_labels=['20k','40k','60k','80k','100k','120k','140k','160k']
    plt.yticks(yloc,y_labels);
    plt.ylabel('total cycling minutes');
    count+=1

fig.suptitle('Cycling Duration Spreading By User_Type for Age_category,Member_Gender and Time of Day');
plt.show();
```

### Spreading by Mean 

```python
variables = ['age_category','member_gender','time_of_day']
count=1
fig, ax = plt.subplots(ncols=len(variables), figsize=[20,8])
for var in variables:
    plt.subplot(1,3,count)
    sb.barplot(data=df_fordgobike, x=var, y='duration_min', hue='user_type');
    plt.xticks(rotation=90)
    plt.ylabel('')
    plt.legend(framealpha=0.5,title='User Type');
    plt.ylabel('total cycling minutes');
    count+=1
fig.suptitle('Mean Spreading By User_Type for Age_category,Member_Gender and Time of Day');
plt.show();
```

## User_Type Average Duration Minutes for Member_Gender, Age_Category and Bike_Share_for_all_trip.

Exploring the trend across all groups

```python
gup_means = df_fordgobike.groupby(['user_type', 'member_gender']).mean()['duration_min']
gup_means = gup_means.reset_index(name = 'duration_min_Avg')
gup_means = gup_means.pivot(index = 'member_gender', columns = 'user_type', values = 'duration_min_Avg')
# sb.heatmap(gup_means, annot =True, fmt = '.3f'
# cbar_kws = {'label': 'mean(num_var2)'})
gup_means

gup_sum = df_fordgobike.groupby(['user_type', 'member_gender']).sum()['duration_min']
gup_sum = gup_sum.reset_index(name = 'duration_min_Avg')
gup_sum = gup_sum.pivot(index = 'member_gender', columns = 'user_type', values = 'duration_min_Avg')
gup_sum


```

```python
fig, ax = plt.subplots(ncols = 2, figsize = [21,9])
labels = np.array([['110k','420k'], ['240k', '1.2mil'], ['12k', '49k']])
sb.heatmap(gup_sum, annot = labels, fmt = '', ax = ax[0], cbar_kws = {'label':'Total cycling duration'});
sb.heatmap(gup_means, annot = True, fmt = '.2f', ax = ax[1], cbar_kws = {'label':'Mean cycling duration'});
fig.suptitle('Total Cycling Time Vs Mean Cycling Time');
plt.show();
```
